### Pertanyaan 1:

Berapa jumlah total rental (cnt) pada musim salju (4) untuk tahun 2012?

sebelum melakukan filter data kita terlebih dahulu cek apakah data yang dimaksud ada

```python
condition_check = (day_df["yr"] == 0) & (day_df["season"] == 1)

# Print the result
print(condition_check.any())
```

```python
# Filter tahun 2012 dan musim salju (season 4)
filtered_data = day_df[(day_df["yr"] == 1) & (day_df["season"] == 4)]

# Hitung jumlah total sewa sepeda (cnt)
total_rental = filtered_data["cnt"].sum()

print("Jumlah rental untuk tahun 2012 selama musim salju(musim 4):", total_rental)
```

Lalu bagaimana untuk musim lainnya?

```python
# Filter tahun 2012 dan musim semi (season 1)
filtered_data = day_df[(day_df["yr"] == 1) & (day_df["season"] == 1)]

# Hitung jumlah total sewa sepeda (cnt)
total_rental = filtered_data["cnt"].sum()

print("Jumlah rental untuk tahun 2012 selama musim semi(musim 1):", total_rental)
```

- musim panas:

```python
# Filter tahun 2012 dan musim panas (season 2)
filtered_data = day_df[(day_df["yr"] == 1) & (day_df["season"] == 2)]

# Hitung jumlah total sewa sepeda (cnt)
total_rental = filtered_data["cnt"].sum()

print("Jumlah rental untuk tahun 2012 selama musim panas(musim 2):", total_rental)
```

```python
# Filter tahun 2012 dan musim gugur (season 3)
filtered_data = day_df[(day_df["yr"] == 1) & (day_df["season"] == 3)]

# Hitung jumlah total sewa sepeda (cnt)
total_rental = filtered_data["cnt"].sum()

print("Jumlah rental untuk tahun 2012 selama musim gugur(musim 3):", total_rental)
```

### Pertanyaan 3:

- Bagaimana Perbandingan Pengguna Teregistrasi dengan Kasual




```python

total_kasual = sum(day_df["casual"])
total_teregistrasi =sum(day_df["registered"])
print("Jumlah Pengguna Kasual adalah: ", total_kasual)
print("Jummlah Pengguna Teregistrasi adalah: ", total_teregistrasi)
```

Untuk memvisualisasikan perbandingan ini alangkah baiknya menggunakan pie plot

```python
import matplotlib.pyplot as plt

# Data
sizes = [total_kasual, total_teregistrasi]  # Ukuran segmen pie plot
labels = ["Pengguna Kasual", "Pengguna Teregistrasi"]  # label untuk segmen pie plot

# Plot
fig, ax = plt.subplots(figsize=(7, 7))
pie = ax.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=90, pctdistance=0.85, shadow=True)

# Customizing colors
colors = ["#FF9999", "#66B2FF"]  # warna untuk setiap segmen
for i, wedge in enumerate(pie[0]):
    wedge.set_edgecolor('white')
    wedge.set_facecolor(colors[i])


plt.setp(pie[2], size=15, weight="bold", color="black")
plt.setp(pie[1], size=12)
plt.title("Perbandingan Tipe Pengguna", fontsize=20)


ax.axis('equal')


plt.show()
```
